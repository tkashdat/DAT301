## Pandas

The purpose of the Python package called Panda is for:

- Working with data sets
- Analyzing cleaning, exploring and manipulating data
- Analyzing big data and making conclusions based on statistical theories
- Cleaning data such as deleting rows or empty values

The name "Pandas" has a reference to both "Panel Data", and "Python Data Analysis" and was created by Wes McKinney in 2008.


Pandas is typically imported as the alias `pd`.

```python
import pandas as pd
```

## Series

A Pandas `Series` is like a column in a table.

It is a one-dimensional array holding data of any type.

Syntax: change a list into a series using `pd.Series()`.

```python
lotr = {
  'elves': ["Elrond", "Arwen", "Legolas", "Galadriel", "Tauriel"],
  'gender': ["M", "F", "M", "F", "F"]
}

lotr_srs = pd.Series(lotr)
print(lotr_srs)
```

### Labels

If nothing else is specified, the values are labeled with their index number. First value has index 0, second value has index 1 etc.

This label can be used to access a specified value.


```python
print(lotr_srs["elves"], '\n') ## print series corresponding to label "elves"
print(lotr_srs["elves"][0])    ## print index 0 from "elves" label
```

### Create Labels
With the `index =` argument, you can name your own labels.  Remember that the labels should be the same length as the series as each row will be named something.

```python
book = ["Fellowship of the Ring", "Two Towers", "Return of the King"]
book_srs = pd.Series(book, index = [1, 2, 3])

print(book_srs, '\n')

print(book_srs[1])  ## reference index 1
```

### Key/Value Objects as Series

You can also use a key/value object, like a dictionary, when creating a Series.  Note: the keys of the dictionary will become the index automatically.

```python
avatar = {"elements": 4, "seasons": 3, "episodes": 54}
avatar_srs = pd.Series(avatar)
print(avatar_srs, '\n')

avatar_srs2 = pd.Series(avatar, index = ["seasons", "episodes"])  ## create a subset of dictionary using index
print(avatar_srs2)
```

## DataFrames
Data sets in Pandas are usually multi-dimensional tables, called DataFrames.

A Pandas DataFrame is a 2 dimensional data structure, like a 2 dimensional array, or a table with rows and columns.

Syntax: create a DataFrame in Pandas using `pd.DataFrame()` function.

```python
chld = {
    "character": ["Sam", "Gerontius", "Fëanor", "Gorbadoc", "Hamfast"],
    "children": [13, 11, 7, 7, 6],
    "species": ["Hobbit", "Hobbit", "Elf", "Hobbit", "Hobbit"]
    
}

chld_df = pd.DataFrame(chld)
print(chld_df)
```

### Locate

Pandas use the `loc` attribute to return one or more specified row(s).

```python
print(chld_df.loc[0], '\n')  ## locate one row

print(chld_df.loc[[0, 2]])   ## locate 2 rows using an array of row references

print(type(chld_df.loc[[0, 2]]))  ## return will be Pandas df when using []
```

### Named Indexes
With the `index =` argument inside `pd.DataFrame()` function, you can name your own indexes.

```python
avatar = {
    "season": ["Water", "Earth", "Fire"],
    "episodes": [20, 20, 21],
    "aired": ["February 21, 2005", "March 17, 2006", "September 21, 2007"]
}

avatar_df = pd.DataFrame(avatar, index = [1, 2, 3])
print(avatar_df)
```

### Locate Named Indexes

Use the named index in the `loc` attribute to return the specified row(s).

```python
avatar_df.loc[[1, 3]]
```

### Load CSV Files Into a DataFrame

If your data sets are stored in a file, Pandas can load them into a DataFrame.

CSV files contains plain text and is a well know format that can be read by everyone including Pandas.

Syntax: `pd.read_csv('data.csv')`

### Viewing the Data

The `df.head(n)` method returns n number of rows, starting from the top, from df.

The `df.tail(n)` method returns n number of rows, starting from the bottom, from df.



```python
print(avatar_df.head(2), '\n')
print(avatar_df.tail(2), '\n')
```

### Info About the Data

The DataFrames object has a method called `info()`, that gives you more information about the data set.

```python
avatar_df.info()
```

### Cleaning Data

### Remove Rows

One way to deal with empty cells is to remove rows that contain empty cells.

Syntax: `df.dropna()`

To remove rows that are not empty use:

Syntax: `df.drop(x, inplace = True)`

### Replace Empty Values

The `df.fillna()` method allows us to replace empty cells with a value.

### Changing Wrong Data

Can change wrong data into the fixed value using `loc` method: `df.loc[7, 'Duration'] = 45` would change the index 7 column "Duration" element to 45.

```python
avtr = {
    "season": ["Water", "Earth", "Fire"],
    "episodes": [20, 20, 20], ## notice episodes for season 3 should be 21
    "aired": ["February 21, 2005", "March 17, 2006", "September 21, 2007"]
}

avtr_df = pd.DataFrame(avtr, index = [1, 2, 3])

avtr_df.loc[3, "episodes"] = 21
print(avtr_df)
```

### Statistics

For statistical analysis example, we will use the Boston data from sklearn package.  The dataset is famous dataset from the 1970’s and used with linear regression analysis.  The data is stored as a dictionary with the response variable MEDV (median value) stored as a "type" and the predictor varaibles stored in "data."  We will first append the data together and then run our analysis on the outputted df.

Note:  There are 506 instances and 14 attributes.

#### Load Boston data from sklearn for our example

```python
from sklearn.datasets import load_boston

boston = load_boston()
print(boston.feature_names, '\n')  ## feature_names will display the names of the columns
print(boston.DESCR)                ## DESCR will provide the dataset characteristics
```

#### Create dataframe

```python
boston_df = pd.DataFrame(boston.data)    ## create df
boston_df.columns = boston.feature_names ## add column names
print(boston_df.head())                  ## notice MEDV is not in this data.  you can access via 'boston.target'


boston_df2 = boston_df.assign(MEDV = boston.target) ## add MEDV
boston_df2.head()
```

#### Specific statistic functions

Syntax: `DataFrame.__()`

May include all of the data in the dataframe or just certain columns pulled using ["", "", ...] method.  By default, axis is index (axis = 0).


- `count()`	Number of non-null observations
- `sum()`	Sum of values
- `mean()`	Mean of Values
- `median()`	Median of Values
- `mode()`	Mode of values
- `std()`	Standard Deviation of the Values
- `min()`	Minimum Value
- `max()`	Maximum Value
- `abs()`	Absolute Value
- `prod()`	Product of Values
- `cumsum()`	Cumulative Sum
- `cumprod()`	Cumulative Product

```python
print(boston_df2["MEDV"].sum(), '\n')
print(boston_df2["MEDV"].mean(), '\n')
print(boston_df2["MEDV"].std(), '\n')
```

#### Finding all statistics

Syntax: `DataFrame.describe(percentiles = None, include = None, exclude = None, datetime_is_numeric = False)`

The describe function will generate descriptive statistics.

Descriptive statistics include those that summarize the central tendency, dispersion and shape of a dataset’s distribution, excluding NaN values.

Analyzes both numeric and object series, as well as DataFrame column sets of mixed data types. The output will vary depending on what is provided. 

```python
boston_df2.describe()
```

Instead of the predefined statistics, specific combinations of aggregating statistics for given columns can be defined using the `DataFrame.agg()` method.  Specify the variable(s) you want the descriptive statistics for as a key and the statistics themselves as values in a list format.

```python
boston_df2.agg({
    "MEDV": ["min", "max", "mean", "median"],
    "TAX": ["min", "max", "mean", "median"]
})
```

#### Aggregating statistics grouped by category
Syntax: `DataFrame.groupby("")`

May be useful when finding mean, median, standard deviation.  Can return all variables or specific column(s).

```python
boston_df2.groupby("CHAS").mean()
```

#### Counts

Syntax: `DataFrae.value_counts()`

```python
boston_df2.groupby("CHAS")["CHAS"].value_counts()
```

#### Correlation

The `corr()` method calculates the relationship between each column in your data set.



```python
boston_df2.corr() ## correlation matrix
```

```python
boston_df2[["MEDV", "TAX"]].corr()
```

### Pandas Plotting

Pandas uses the `plot()` method to create diagrams.

Pythons uses `Pyplot`, a submodule of the Matplotlib library to visualize the diagram on the screen.


```python
import matplotlib.pyplot as plt
```

### Scatter Plot
Specify that you want a scatter plot with the kind argument:

`kind = 'scatter'`

A scatter plot needs an x- and a y-axis.

Include the x and y arguments like this:

`x = ''`, `y = ''`

Syntax: `df.plot(kind = "scatter", x = "", y = "", ...)`

#### Draw scatterplot

Draw a scatterplot comparing the TAX variable to the LSTAT variable.

```python
boston_df2.plot(kind = "scatter", x = "TAX", y = "MEDV")
plt.suptitle("Boston Housing Prices")
plt.show()
```
