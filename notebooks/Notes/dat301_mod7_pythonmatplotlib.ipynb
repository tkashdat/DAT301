## Matplotlib

Matplotlib is a low level graph plotting library in python that serves as a visualization utility.

Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias:

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

## Line Graphs

```python
## draw a line in plt

xpoints = np.array([0, 5])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints)
plt.show()
```

### Plotting x and y points

The `plot()` function is used to draw points (markers) in a diagram.  

A line plot is default.

The function takes parameters for specifying points in the diagram.

Parameter 1 is an array containing the points on the x-axis.

Parameter 2 is an array containing the points on the y-axis.

#### Example:
Draw a line connecting points (1, 3) to (8, 10).  Notice you need to pass two arrays [1, 8] and [3, 10] to create the plot.

```python
xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

#### Example:
Draw a line in a diagram from positions (1, 3) to (2, 8) and finally position (8, 10).

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

### Markers
Add markers with `marker =` option.


- `o` is a point
<br>

- `*` is a star
<br>

- `x` is an x 
<br>
...

```python
plt.plot(xpoints, ypoints, marker = "*")
plt.show()
```

### Formatting
Format strings with fmt written with this syntax: `marker|line|color`

- lines include: "-" solid line, ":" dotted line, "--" dashed line, "-." dashed dotted line
<br>

- colors include: "r" for red, "g" for green", "b" for blue, "k" for black, "w" for white




```python
plt.plot(xpoints, ypoints, "*:g")
plt.show()
```

- format marker size with `ms` = integer
<br>

- format marker edge color with `mec` = "r" for red, "g" for green, "b" for blue, "k" for black, "w" for white
<br>

- format marker face color with `mfc` = "r" for red, "g", for green, "b" for blue, "k" for black, "w" for white

```python
plt.plot(xpoints, ypoints, "*-g", ms = 10, mec = "y", mfc = "w")
plt.show()
```

### Linestyle
You can use the keyword argument `linestyle`, or shorter `ls`, to change the style of the plotted line.

Format options include:

- `solid` or `-` (default)
<br>

- `dotted` or `:`
<br>

- `dashed` or `--`
<br>

- `dashdot` or `-.`
<br>

- `None`, `''`, or `' '`


```python
plt.plot(xpoints, ypoints, "*:g", ls = ":", ms = 10, mec = "y", mfc = "w")
plt.show()
```

### Line Color
You can use the keyword argument `color` or the shorter `c` to set the color of the line.

You can use hexadecimal color values.

1 of the 140 colors (see: https://www.w3schools.com/colors/colors_names.asp)

Or typical "r" for red, "g" for green, "b" for blue, "k" for black, "w" for white.

### Line Width
You can use the keyword argument `linewidth` or the shorter `lw` to change the width of the line.

The value is a floating number, in points.

#### Example:

```python
plt.plot(xpoints, ypoints, ls = ":", c = "Chartreuse", lw = 6)
plt.show()
```

### Multiple Lines
You can plot as many lines as you like by simply adding more `plt.plot()` functions in the same code chunk.

You can also plot many lines by adding the points for the x- and y-axis for each line in the same `plt.plot()` function.

#### Example
Adding two `plt.plot()` functions.

```python
x1 = np.array([1, 2, 3, 4, 5])
y1 = np.array([.1, .05, .8, .05, 0])
y2 = np.array([.05, .25, .45, .2, .05])

plt.plot(x1, y1, ls = ":", c = "r", lw = 2)
plt.plot(x1, y2, ls = "--", c = "k", lw = 2)
plt.show()
```

#### Example
Adding two plots ot one `plt.plot()` function.

```python
x1 = np.array([1, 2, 3, 4, 5])
y1 = np.array([.1, .05, .8, .05, 0])
y2 = np.array([.05, .25, .45, .2, .05])

plt.plot(x1, y1, "*:g", x1, y2, "o--k")
plt.show()
```

### Display Multiple Plots
With the `subplots()` function you can draw multiple plots in one figure.

The `subplots()` function takes three arguments that describes the layout of the figure.

Syntax:
<br>
`subplots(row, column, index)`
<br>

If we want a figure with 2 rows and 1 column (meaning that the two plots will be displayed on top of each other instead of side-by-side), we can write the syntax like this:
<br>
`subplot(2, 1, 1)` ## first graph
`subplot(2, 1, 2)` ## second graph

For side-by-side:
<br>
`subplot(1, 2, 1)` ## first graph
`subplot(1, 2, 2)` ## second graph

The plot formatting can be added as above.

#### Example

```python
#plot 1:
x = range(1, 21, 1)
y = np.log(x)

plt.subplot(1, 2, 1)
plt.plot(x, y, "*--r")
plt.title("Log Graph")

#plot 2:
x = range(1, 20, 2)
y = range(0, 30, 3)

plt.subplot(1, 2, 2)
plt.plot(x, y, "o-.g")
plt.title("Linear Graph")

plt.suptitle("Graphs") ## Add overall title
plt.show()
```

## Scatterplots

With Pyplot, you can use the `scatter()` function to draw a scatter plot.

The `scatter()` function plots one dot for each observation. It needs two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis.

### Formatting

- You can set your own color for each scatter plot with the `color =` or the `c =` argument.
<br>

- You can set a specific color for each dot by using an array of colors as value for the `c =` (not `color =`) argument.
<br>


```python
x = np.array([2, 3, 4, 5, 6, 7, 8, 9, 10])
y = np.array([3, 10, 4, 11, 5, 12, 6, 13, 7])
colors = np.array(["red", "green", "blue", "yellow", "pink", "black", "orange", "purple", "beige"])

plt.scatter(x, y, c = colors)
plt.show()
```

### ColorMap

The Matplotlib module has a number of available colormaps.

A colormap is like a list of colors, where each color has a value that ranges from 0 to 100.

Syntax: 
<br>
`cmap =` the desired value of the colormap.

Built in colormaps (see: https://matplotlib.org/stable/gallery/color/colormap_reference.html):

- Accent
- BrBG
- BuGn
- BuPu
- CMRmap
- Dark2
- GnBu
- Greens
- Greys
- OrRd
- Oranges
- ...

You can display the colormap legend using `plt.colorbar()`.

#### Example

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

cs = np.array([0, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60])

plt.scatter(x, y, c = cs, cmap = 'viridis')
plt.colorbar()
plt.show()
```

### Size
You can change the size of the dots with the `s` argument.

Just like colors, make sure the array for sizes has the same length as the arrays for the x- and y-axis.

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
cs = np.array([0, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60])

plt.scatter(x, y, c = cs, cmap = "winter", s = sizes)
plt.colorbar()
plt.show()
```

### Alpha
You can adjust the transparency of the dots with the alpha argument.

Just like colors, make sure the array for sizes has the same length as the arrays for the x- and y-axis.

```python
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c = colors, s = sizes, alpha = .5, cmap = 'twilight')

plt.colorbar()

plt.show()
```

#### Example
Using diabetes data (see: https://www4.stat.ncsu.edu/~boos/var.select/diabetes.html) plot blood sugar levels (10th column variable) against disease progression ("target").  Colorize based on bmi (column 3).

Note:  Each of these 10 feature variables have been mean centered and scaled by the standard deviation times n_samples (i.e. the sum of squares of each column totals 1).

```python
import sklearn
from sklearn import datasets
import matplotlib.pyplot as plt
import numpy as np

diabetes = datasets.load_diabetes()
x = diabetes.data[:, 9] ## load sugar variable
col = diabetes.data[:, 2] ## load bmi variable
y = diabetes["target"] ## load y value

plt.scatter(x, y, c = col, alpha = 0.5, cmap = 'nipy_spectral')
cbar = plt.colorbar() ## plot colorbar but name it so you can add a label in the below code line
cbar.set_label("BMI")
plt.xlabel(diabetes.feature_names[9])
plt.ylabel("Disease Progression")
plt.suptitle("Diabetes Disease Progression")
plt.show()
```

```python
import sklearn
from sklearn import datasets
import matplotlib.pyplot as plt
import numpy as np

iris = datasets.load_iris()
X_iris = iris.data[:, :2]  ## only take the first two features from columns, all rows
Y_iris = iris.target
n_classes = 3

for i in range(n_classes):
    index = np.where(Y_iris == i)
    plt.scatter(X_iris[index, 0], X_iris[index, 1],  
    label=iris.target_names[i])
plt.legend()
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.suptitle("Iris Dataset")
plt.show()
```

## Histograms

A histogram is used to give an approximate representation of the distribution of the data, based on the sample data at hand. A histogram is constructed by using equally sized bins (intervals), and counting the number of data points that belong to each bin. Creating histograms at the start of a new project is very useful to get familiar with the data, and to get a rough sense of the density of the underlying distribution.

To create a histogram, we make use of the `plt.hist()` function.

MatPlotLib's `hist()` function will use an array of numbers to create a histogram.  The `hist()` function will read the array and produce a histogram.

```python
## we will use the california housing data below to create histograms.  first import the
## data from the sklearn datasets package and then change the data into a dataframe

from sklearn.datasets import fetch_california_housing
cal_housing = fetch_california_housing(as_frame = True)
hous = cal_housing['data'].join(cal_housing['target'])
hous
```

```python
plt.hist(hous["HouseAge"])
```

### Format

- Change number of bins with `bins =` argument inside `plt.hist()` function (default is 10)
<br>

- Use `plt.title()` to set a title
<br>

- Use `plt.xlabel()` and `plt.ylabel()` to change axis labels
<br>

- Use `color =` inside `plt.hist()` to change color.  Can have 1 color per dataset.  All 140 color names are supported (see: https://www.w3schools.com/colors/colors_names.asp) as are hexidecimal color system.
<br>

- Use `density = True` to show the density (relative frequency on y axis).  Default is False

```python
bins = 20

plt.hist(hous["HouseAge"], bins, color = "darkslategrey", density = True) # Show realtive density!!
plt.ylabel("Density")
plt.xlabel("Home Age")
plt.title("California Home Age Distribution")
```

Instead of plotting the histogram for a single feature, we can plot the histograms for all features. This can be done by creating separate plots, but here, we will make use of subplots, so that all histograms are shown in one single plot. For this, we make use of the `plt.subplots()` function. By using the following code, we obtain the plot containing the four histograms:

```python
fig, axs = plt.subplots(2, 2)
axs[0, 0].hist(hous["HouseAge"])
axs[0, 0].set_title("House Age")
axs[0, 1].hist(hous["MedHouseVal"], color = "orange")
axs[0, 1].set_title("Median House Value")
axs[1, 0].hist(hous["MedInc"], color = "green")
axs[1, 0].set_title("Median Income")
axs[1, 1].hist(hous["Population"], color = "red")
axs[1, 1].set_title("Population")

for ax in axs.flat: # use axs.plot to iterate through all plots in 2D plot grid
    ax.set(ylabel = 'frequency')

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()

fig.suptitle("California Housing Data")
plt.show()
```

## Pie Charts

With Pyplot, you can use the `pie()` function to draw pie charts.

The pie chart draws one piece (called a wedge) for each value in the array.

Note: The size of each wedge is determined by comparing the value with all the other values, by using this formula:

The value divided by the sum of all values: x/sum(x).

By default the plotting of the first wedge starts from the x-axis and move counterclockwise.  You can change the start angle by specifying a startangle parameter.  The startangle parameter is defined with an angle in degrees, default angle is 0

```python
## LOTR data

lotr_count = np.array([410+62, 78+20, 9+3, 172+74, 51, 27+10, 20])
lotr_spec = np.array(["Men", "Elves", "Half-Elves", "Hobbits", "Dwarves", "Ainur", "Orcs"])
```

```python
plt.pie(lotr_count)
plt.show()
```

### Labels
Add labels to the pie chart with the `labels =` parameter.

The label parameter must be an array with one label for each wedge.

```python
plt.pie(lotr_count, labels = lotr_spec, startangle = 90)
plt.show()
```

### Explode
Maybe you want one of the wedges to stand out? The `explode =` parameter allows you to do that inside of the `plt.pie()` function.

The explode parameter, if specified, and not None, must be an array with one value for each wedge.

Each value represents how far from the center each wedge is displayed.

```python
ex = [0, 0.2, 0, 0, 0, 0, 0]

plt.pie(lotr_count, labels = lotr_spec, startangle = 90, explode = ex)
plt.show()
```

### Colors
You can set the color of each wedge with the `colors =` parameter inside the `plt.pie()` function.

The colors parameter, if specified, must be an array with one value for each wedge.

### Legend
To add a list of explanation for each wedge, use the `legend()` function.

To add a header to the legend, add the `title =` parameter inside the legend function.

To change position of legend so it isn't overlapping, use `bbox_to_anch =` (specify the (x,y) coords) and `loc =` parameters.

```python
ex = [0, 0.2, 0, 0, 0, 0, 0]
colors = ["green", "purple", "lavender", "red", "orange", "beige", "black"]

plt.pie(lotr_count, labels = lotr_spec, startangle = 90, explode = ex, colors = colors)
plt.legend(bbox_to_anchor=(1.2, 1.025), loc = "upper left")
plt.show()
```
